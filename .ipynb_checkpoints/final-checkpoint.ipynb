```python
import pandas as pd
```

```python
df=pd.read_csv("house_price.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df=df.drop(columns=["date","country","statezip","street","city"],axis=1)
```

```python
df["price"]=df["price"].astype(int)
```

```python
df=df[(df["price"]>0) & (df["price"]<2000000)]
```

```python
df
```

```python
df.dtypes
```

```python
X=df.drop(columns=["price"],axis=1)
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr_pred=lr.predict(X_test)
```

```python
print("Score of Training in LR: ",lr.score(X_train,y_train))
print("Score of Testing in LR: ",lr.score(X_test,y_test))
```

```python
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse
import numpy as np
```

```python
print("Mean Absolute error in LR: ",mae(y_test,lr_pred))
print("Mean Squared error in LR: ",mse(y_test,lr_pred))
print("Root means squared error in LR: ",np.sqrt(mse(y_test,lr_pred)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(
    n_estimators=300,
    max_depth=None,
    min_samples_split=5,
    random_state=42
)
rfr.fit(X_train,y_train)
```

```python
rfr_pred=rfr.predict(X_test)
```

```python
print("Score of Training in rfr: ",rfr.score(X_train,y_train))
print("Score of Testing in rfr: ",rfr.score(X_test,y_test))
```

```python
print("Mean Absolute error in rfr: ",mae(y_test,rfr_pred))
print("Mean Squared error in rfr: ",mse(y_test,rfr_pred))
print("Root means squared error in rfr: ",np.sqrt(mse(y_test,rfr_pred)))
```

```python
from sklearn.tree import DecisionTreeRegressor

dtr=DecisionTreeRegressor()
dtr.fit(X_train,y_train)
```

```python
dtr_pred=dtr.predict(X_test)
```

```python
print("Score in Training DTR:",dtr.score(X_train,y_train))
print("Score in Testing DTR: ",dtr.score(X_test,y_test))

print("Mean Absolute Error: ",mae(y_test,dtr_pred))
print("Mean Squaredd Error: ",mse(y_test,dtr_pred))
print("Root mean Squared Error: ",np.sqrt(mse(y_test,dtr_pred)))
```

```python
from sklearn.svm import SVR

svr=SVR()
svr.fit(X_train,y_train)
```

```python
svr_pred=svr.predict(X_test)
```

```python
print("Score in Training SVR:",svr.score(X_train,y_train))
print("Score in Testing SVR: ",svr.score(X_test,y_test))

print("Mean Absolute Error: ",mae(y_test,svr_pred))
print("Mean Squaredd Error: ",mse(y_test,svr_pred))
print("Root mean Squared Error: ",np.sqrt(mse(y_test,svr_pred)))
```

```python
import numpy as np

def predict_house_price(
    bedrooms=3,
    bathrooms=2,
    sqft_living=1500,
    sqft_lot=6000,
    floors=1,
    waterfront=0,
    view=0,
    condition=3,
    sqft_above=1500,
    sqft_basement=0,
    yr_built=1990,
    yr_renovated=0
):

    temp_array = [
        bedrooms,
        bathrooms,
        sqft_living,
        sqft_lot,
        floors,
        waterfront,
        view,
        condition,
        sqft_above,
        sqft_basement,
        yr_built,
        yr_renovated
    ]

    temp_array = np.array([temp_array])

    prediction = lr.predict(temp_array)

    return int(prediction[0])
```

```python
price = predict_house_price(
    bedrooms=3,
    bathrooms=1,
    sqft_living=1090,
    sqft_lot=6005,
    floors=1,
    waterfront=0,
    view=0,
    condition=4,
    sqft_above=1090,
    sqft_basement=0,
    yr_built=1955,
    yr_renovated=2009
)

print("Predicted Price:", price)
```
